```python
import pandas as pd

dfs = []

for path in json_files:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)

        temp_df = pd.json_normalize(data)
        temp_df["source_file"] = path
        dfs.append(temp_df)

    except Exception as e:
        print("로드 실패:", path, e)

df = pd.concat(dfs, ignore_index=True)

print("데이터 개수:", len(df))
df.head()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("c")
```

```python
import chromadb

DB_PATH = "/content/drive/MyDrive/chroma_problem_db"

client = chromadb.PersistentClient(path=DB_PATH)

collection = client.get_or_create_collection(
    name="problem_collection"
)
```

```python
metadata_cols = [
    "raw_data_info.school",
    "raw_data_info.grade",
    "raw_data_info.semester",
    "raw_data_info.subject",
    "source_data_info.level_of_difficulty",
    "source_data_info.types_of_problems",
    "source_data_info.2009_achievement_standard",
    "source_data_info.2015_achievement_standard",
    "source_data_info.2022_achievement_standard"
]

available_metadata_cols = [
    col for col in metadata_cols
    if col in df.columns
]

texts = df["text"].tolist()
ids = [f"problem_{i}" for i in range(len(texts))]

metadatas = []

for i, row in df.iterrows():
    metadata = {
        "source_file": str(row.get("source_file", ""))
    }

    for col in available_metadata_cols:
        metadata[col] = str(row.get(col, ""))

    metadatas.append(metadata)

print("문서 수:", len(texts))
print("메타데이터 컬럼:", available_metadata_cols)
```

```python
from tqdm import tqdm

batch_size = 100

for start in tqdm(range(0, len(texts), batch_size)):
    end = start + batch_size

    batch_texts = texts[start:end]
    batch_ids = ids[start:end]
    batch_metadatas = metadatas[start:end]

    batch_embeddings = model.encode(
        batch_texts,
        show_progress_bar=False
    ).tolist()

    collection.add(
        ids=batch_ids,
        documents=batch_texts,
        embeddings=batch_embeddings,
        metadatas=batch_metadatas
    )
```

```python
print("ChromaDB 저장 개수:", collection.count())
print("DB 저장 위치:", DB_PATH)
```

```python
query = "초등학교 수학 문제"

query_embedding = model.encode([query]).tolist()

results = collection.query(
    query_embeddings=query_embedding,
    n_results=5
)

results
```
